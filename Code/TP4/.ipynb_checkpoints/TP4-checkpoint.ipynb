```python
# Limpieza

def eliminar_categoria_9(df, column_list):
    """
    Elimina la categoría con valor 9 de una serie de variables en un DataFrame.
    
    Parámetros:
    - df: DataFrame donde se encuentran las columnas.
    - column_list: Lista de columnas en las que se quiere eliminar la categoría con valor 9.
    
    Retorna:
    - DataFrame con las categorías con valor 9 eliminadas en las columnas especificadas.
    """
    for column in column_list:
        df = df[df[column] != 9]
    return df


df3 = eliminar_categoria_9(df2, ['NIVEL_ED', 'CH08', 'CH07', 'CAT_OCUP', 'CH15', 'CH16']) 

# Edad menor a cero eliminada
df4 = df3[(df3['CH06'] > 0)] 
```

```python
def columnas_duplicadas(df):
    """
    Esta función verifica si hay columnas con valores idénticos en un DataFrame.
    
    Argumentos:
    - df: DataFrame de pandas
    
    Retorna:
    - Una lista de tuplas con los nombres de las columnas repetidas.
    """
    
    columnas_repetidas = []
    
    for i in range(df.shape[1]):
        for j in range(i+1, df.shape[1]):
            if df.iloc[:, i].equals(df.iloc[:, j]):
                columnas_repetidas.append((df.columns[i], df.columns[j]))
    
    return columnas_repetidas


def eliminar_columnas_duplicadas(df):
    # Obtener las columnas duplicadas
    cols_duplicadas = columnas_duplicadas(df)
    
    # Crear una lista de columnas para eliminar
    columnas_para_eliminar = [col[1] for col in cols_duplicadas]
    
    # Eliminar las columnas
    df = df.drop(columns=columnas_para_eliminar)
    
    return df

eliminar_columnas_duplicadas(df4) # verificamos si hubiera columnas duplicadas y las eliminamos
```

```python
def drop_high_na(df, threshold=0.7):
    """
    Elimina las columnas y las filas del DataFrame que tienen un porcentaje de valores faltantes 
    superior al umbral especificado.

    Parámetros:
    - df: DataFrame de entrada.
    - threshold: umbral de porcentaje de valores faltantes para eliminar una columna o fila (valor predeterminado = 0.7).

    Retorna:
    - DataFrame después de eliminar las columnas y filas con un alto porcentaje de valores faltantes.
    """
    # Calcula el porcentaje de valores faltantes para cada columna
    missing_percentage_col = df.isnull().mean()
    
    # Filtra las columnas que tienen un porcentaje de valores faltantes superior al umbral
    columns_to_drop = missing_percentage_col[missing_percentage_col > threshold].index
    
    # Elimina las columnas seleccionadas del DataFrame
    df_cleaned = df.drop(columns=columns_to_drop)
    
    # Calcula el porcentaje de valores faltantes para cada fila
    missing_percentage_row = df_cleaned.isnull().mean(axis=1)
    
    # Filtra las filas que tienen un porcentaje de valores faltantes superior al umbral
    rows_to_drop = missing_percentage_row[missing_percentage_row > threshold].index
    
    # Elimina las filas seleccionadas del DataFrame
    df_cleaned = df_cleaned.drop(index=rows_to_drop)
    
    return df_cleaned

drop_high_na(df4)
```

### 1) Escriban una función, llamada evalua metodo, que reciba como argumentos un modelo y los datos de entrenamiento y prueba (X train, y train, X test, y test). La función debe ajustar el modelo con los datos de entrenamiento y calcular las métricas que considere necesarias para esta problemática (de mínima, deben reportar verdaderos positivos, verdaderos negativos, falsos positivos, falsos negativos, AUC, accuracy y precision de cada método). El output de la función debe ser una colección con las métricas evaluadas.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler

# Separar las características (X) y la variable dependiente (y)
X = respondieron.drop('pobre', axis=1)
X = X.select_dtypes(include=["number"])  # Mantenemos las variables numéricas

# Función para aplicar imputación diferente según el número de valores únicos
def impute_column(column):
    if len(column.unique()) > 2:
        return column.fillna(column.mean())
    else:
        return column.fillna(column.median())

# Aplicar la imputación personalizada a cada columna
X_imputed = X.apply(impute_column, axis=0)

# Crear y aplicar escalador robusto para manejar outliers
escalador_robusto = RobustScaler()
X_scaled = pd.DataFrame(escalador_robusto.fit_transform(X_imputed), columns=X_imputed.columns, index=X_imputed.index)

# Asegurarse de que y tenga las mismas filas que X_scaled
y_cleaned = respondieron.loc[X_scaled.index, 'pobre']

# Dividir los datos en conjuntos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(X_scaled, y_cleaned, test_size=0.3, random_state=2023)

print("Conjunto de entrenamiento:", "X_train =", len(x_train), "Y_train =", len(y_train))
print("Conjunto de prueba:", "X_test =", len(x_test), "Y_test =", len(y_test))
```

```python
from sklearn.metrics import confusion_matrix, roc_auc_score, accuracy_score, mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier


def evalua_modelo(x_train, y_train, x_test, y_test, tipos=['logistico', 'LDA', 'kNN', 'RF', 'DecisionTree', 'Bagging', 'GradientBoosting'], 
                  lambda_=1.0, k_neighbors=5, penalidad='l2', n_estimators=100, max_features='sqrt', max_depth=None, n_bagging_estimators=10, learning_rate=0.1):
    """
    Evalúa varios modelos de clasificación y retorna un DataFrame con métricas de rendimiento.

    Parámetros:
    - x_train, y_train: características y objetivo del conjunto de entrenamiento.
    - x_test, y_test: características y objetivo del conjunto de prueba.
    - tipos: lista de modelos a evaluar. Opciones: 'logistico', 'LDA', 'kNN', 'RF'.
    - lambda_: parámetro de regularización para la regresión logística.
    - k_neighbors: número de vecinos para kNN.
    - penalidad: tipo de penalidad para la regresión logística ('l1' o 'l2').
    - n_estimators: número de árboles para Random Forest.
    - max_features: número de características a considerar para cada árbol en Random Forest.
    
    Retorna:
    - DataFrame con métricas de rendimiento (AUC, Accuracy, ECM, True Positives, False Positives, True Negatives, False Negatives) para cada modelo.
    """

    x_train = np.ascontiguousarray(x_train.values if isinstance(x_train, pd.DataFrame) else x_train)
    x_test = np.ascontiguousarray(x_test.values if isinstance(x_test, pd.DataFrame) else x_test)

    resultados = pd.DataFrame({
        'Metrica': ['AUC', 'Accuracy', 'ECM', 'True P', 'False P', 'True N', 'False N'],
        'Lambda': [np.nan] * 7,  # Columna vacía para lambda
        'k_neighbors': [np.nan] * 7,  # Columna vacía para k_neighbors
        'Penalidad': [np.nan] * 7,  # Columna vacía para penalty
        'n_estimators': [np.nan] * 7,  # Columna vacía para n_estimators
        'max_features': [np.nan] * 7  # Columna vacía para max_features
    })

    for tipo in tipos:
        if tipo == 'logistico':
            solver_choice = 'liblinear' if penalidad == 'l1' else 'lbfgs'
            modelo = LogisticRegression(C=1/lambda_, penalty=penalidad, solver=solver_choice, max_iter=1000).fit(x_train, y_train)
            resultados['Lambda'] = lambda_
            resultados['Penalidad'] = penalidad
        elif tipo == 'LDA':
            modelo = LinearDiscriminantAnalysis(solver='lsqr', shrinkage='auto').fit(x_train, y_train)
        elif tipo == 'kNN':
            modelo = KNeighborsClassifier(n_neighbors=k_neighbors).fit(x_train, y_train)
            resultados['k_neighbors'] = k_neighbors
        elif tipo == 'RF':
            modelo = RandomForestClassifier(n_estimators=n_estimators, max_features='sqrt', random_state=42).fit(x_train, y_train)
            resultados['n_estimators'] = n_estimators
            resultados['max_features'] = max_features
        elif tipo == 'DecisionTree':
            modelo = DecisionTreeClassifier(max_depth=max_depth, random_state=42).fit(x_train, y_train)
        elif tipo == 'Bagging':
            modelo_base = DecisionTreeClassifier(max_depth=max_depth)
            modelo = BaggingClassifier(estimator=modelo_base, n_estimators=n_bagging_estimators, random_state=42).fit(x_train, y_train)
        elif tipo == 'GradientBoosting':
            modelo = GradientBoostingClassifier(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth, random_state=42).fit(x_train, y_train)
        else:
            raise ValueError(f"Tipo de modelo {tipo} no reconocido.")
        
        y_pred = modelo.predict(x_test)
        y_pred_proba = modelo.predict_proba(x_test)[:, 1] if tipo != 'LDA' else modelo.decision_function(x_test)
        
        conf_matrix = confusion_matrix(y_test, y_pred)
        auc = roc_auc_score(y_test, y_pred_proba)
        acc = accuracy_score(y_test, y_pred)
        ecm = mean_squared_error(y_test, y_pred)
        
        resultados[f'Valor_{tipo}'] = [auc, acc, ecm, 
                                      conf_matrix[1, 1], conf_matrix[0, 1],
                                      conf_matrix[0, 0], conf_matrix[1, 0]]

    for col in resultados.columns[3:]:
        resultados[col] = resultados[col].apply(lambda x: round(x, 2) if isinstance(x, (int, float)) else x)

    return resultados
```

```python
from sklearn.exceptions import ConvergenceWarning

warnings.filterwarnings("ignore", category=ConvergenceWarning) 
# No mostramos el alcance al límite de iteraciones para optimizar el tiempo de trabajo y 
# hacer más rápida la lectura del código 


evalua_modelo(x_train, y_train, x_test, y_test)
```

### 2) Escriban una funcion, llamada cross validation, que realice validacion cruzada con k iteraciones (k-fold CV), llamando a la funcion del inciso anterior en cada una, pero para las k distintas particiones. La funcion debe recibir como argumentos el modelo, el valor de k y un dataset (es decir, solo X e y1). Pueden ayudarse con la funcion KFold para generar las particiones necesarias

```python
from sklearn.model_selection import StratifiedKFold

def cross_validation(X, y, k=5, modelos=['logistico', 'LDA', 'kNN', 'RF', 'DecisionTree', 'Bagging', 'GradientBoosting'], 
                     penalidad='l2', lambda_=1.0, k_neighbors=5, n_estimators=100, 
                     max_features='sqrt', max_depth=None, n_bagging_estimators=10, learning_rate=0.1):
    """
    Realiza validación cruzada para varios modelos de clasificación.

    Parámetros:
    - X, y: características y objetivo del conjunto de datos.
    - k: número de particiones para la validación cruzada.
    - modelos: lista de modelos a evaluar.
    - penalidad: tipo de penalidad para la regresión logística.
    - lambda_: parámetro de regularización para la regresión logística.
    - k_neighbors: número de vecinos para kNN.
    - n_estimators: número de árboles para modelos basados en árboles.
    - max_features: número de características a considerar para cada árbol en modelos basados en árboles.
    - max_depth: profundidad máxima para árboles.
    - n_bagging_estimators: número de estimadores para Bagging.
    - learning_rate: tasa de aprendizaje para Gradient Boosting.

    Retorna:
    - DataFrame con métricas de rendimiento de cada modelo en cada partición.
    """

    skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=2023)
    resultados_totales = []

    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        resultados = evalua_modelo(X_train, y_train, X_test, y_test, tipos=modelos, 
                                   penalidad=penalidad, lambda_=lambda_, k_neighbors=k_neighbors, 
                                   n_estimators=n_estimators, max_features=max_features, 
                                   max_depth=max_depth, n_bagging_estimators=n_bagging_estimators, 
                                   learning_rate=learning_rate)
        resultados_totales.append(resultados)

    resultados_df = pd.concat(resultados_totales, axis=0).reset_index(drop=True)
    resultados_df['Fold'] = [i+1 for i in range(k) for _ in range(len(resultados))]
    
    cols = resultados_df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    resultados_df = resultados_df[cols]

    return resultados_df
```

### 3) Escriban una funcion, llamada evalua config que reciba una lista de configuraciones de hiperparametros (los distintos valores a probar como hiperparametros podrian codificarse en diccionarios de Python) y utilizando la funcion cross validation obtenga el error2 promedio para cada configuracion. Finalmente, la funcion debe devolver la configuracion que genere menor error. Aseg´urense de que esta funcion sirva para cualquier hiperparametro que quieran elegir por crossvalidation para cualquier modelo.

```python
# 3) 

import random

def evalua_config(X, y, lambda_values=[1.0], k_neighbors_values=[5], penalidades=['l2'], n_estimators_values=[100], 
                 max_features_values=['sqrt'], max_depth_values=[None], n_bagging_estimators_values=[10], 
                 learning_rate_values=[0.1], k=5, modelos=['logistico', 'LDA', 'kNN', 'RF', 'DecisionTree', 'Bagging', 'GradientBoosting']):
    """
    Evalúa diferentes configuraciones de hiperparámetros utilizando validación cruzada.

    Parámetros:
    - X, y: características y objetivo del conjunto de datos.
    - lambda_values, k_neighbors_values, etc.: rangos de valores de hiperparámetros a evaluar.
    - k: número de particiones para la validación cruzada.
    - modelos: lista de modelos a evaluar.

    Devuelve:
    - Tuple de (mejor_error, mejor_configuracion)
    """
    mejor_error = float('inf')
    mejor_configuracion = None
    num_configuraciones = 10  # Número de configuraciones aleatorias a probar

    for _ in range(num_configuraciones):
        config = {
            'lambda_': random.choice(lambda_values),
            'k_neighbors': random.choice(k_neighbors_values),
            'penalidad': random.choice(penalidades),
            'n_estimators': random.choice(n_estimators_values),
            'max_features': random.choice(max_features_values),
            'max_depth': random.choice(max_depth_values),
            'n_bagging_estimators': random.choice(n_bagging_estimators_values),
            'learning_rate': random.choice(learning_rate_values)
        }

        resultados_df = cross_validation(X, y, k=k, modelos=modelos, **config)

        for metodo in modelos:
            columna_ecm = f'Valor_{metodo}'
            if columna_ecm in resultados_df.columns:
                ecm_promedio = resultados_df[resultados_df['Metrica'] == 'ECM'][columna_ecm].mean()
                
                if ecm_promedio < mejor_error:
                    mejor_error = ecm_promedio
                    mejor_configuracion = config
                    mejor_configuracion['metodo'] = metodo  # Agregar información sobre qué método produjo este error

    return mejor_error, mejor_configuracion
```
